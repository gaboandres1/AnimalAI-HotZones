```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
def get_train_data(modelname):
    data = pd.read_csv("./summaries/"+modelname+"_AnimalAI.csv")
    steps = data['Steps']
    lesson = data['Environment/Lesson']
    reward = data['Environment/Cumulative Reward']
    length = data['Environment/Episode Length']

    lesson_tr = lesson.astype(float)
    for i, tr in enumerate(tresholds):
        actual = np.where(lesson==i)[0]
        lesson_tr[actual] = tr*np.ones(len(actual))
    return [steps, lesson, reward, length, lesson_tr]
```

```python
def plot_train_data(results, modelname, save=True):
    [steps, lesson, reward, length, lesson_tr] = results
    plt.figure(figsize=(13,5))
    plt.plot(steps, reward, label='Episode reward')
    plt.plot(steps, lesson_tr, label='Threshold')
    plt.xlabel('Step')
    plt.ylabel('Reward')
    plt.title('Environment/Cumulative Reward: '+modelname)
    plt.legend()
    plt.grid()
    if save:
        plt.savefig('./graphs/'+modelname+'_reward.png')
    plt.show()

    plt.figure(figsize=(13,5))
    plt.subplot(121)
    plt.plot(steps, length)
    plt.xlabel('Step')
    plt.ylabel('time')
    plt.title('Environment/Episode Length')
    plt.grid()
    plt.subplot(122)
    plt.plot(steps, lesson)
    plt.xlabel('Step')
    plt.ylabel('Episode')
    plt.title('Environment/Lesson')
    plt.suptitle(modelname)
    plt.grid()
    if save:
        plt.savefig('./graphs/'+modelname+'_EC.png')
    plt.show()
```

Los modelos corresponden a los llamados hotzone_ppoX.
